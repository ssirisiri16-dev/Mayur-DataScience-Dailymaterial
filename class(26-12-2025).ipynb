```python
import math
def construct_cdf(data):
    lst=[]
    mean=np.mean(data)
    sd=np.std(data)
    for i in data:
        s=(i-mean)/(math.sqrt(2)*sd)
        cdf_value=0.5*(1+math.erf(s))
        lst.append(cdf_value)
    return lst
    
    
    
```

```python
data=df["sepal_length"]
cdf_data=construct_cdf(data)
```

```python
cdf_data
```

```python
plt.plot(data,cdf_data)
plt.show()
```
