```python
def permutation_test(sample1,sample2,no_of_samples):
    original_stats=np.mean(sample1)-np.mean(sample2)
    combined_data=np.concatenate([sample1,sample2])
    diff_lst=[]
    for i in range(no_of_samples):
        np.random.shuffle(combined_data)
        s1=combined_data[:len(sample1)]
        s2=combined_data[len(sample1):len(sample1)+len(sample2)]
        s=np.mean(s1)-np.mean(s2)
        diff_lst.append(s)
    diff_lst=np.array(diff_lst)
    sorted_diff_lst=np.sort(diff_lst)
    p_value=(np.sum(sorted_diff_lst>original_stats)/no_of_samples)*100
    return p_value       
```

```python
sample1=np.random.randn(50)
sample2=np.random.randn(50)
no_of_samples=1000
permutation_test(sample1,sample2,no_of_samples)
```
