```python
def min_coins(lst,amount):
    lst.sort(reverse=True)
    result=0
    for i in range(0,len(lst)):
        if lst[i]<=amount:
            count=amount//lst[i]
            result+=count
            amount=amount-count*lst[i]
        if amount==0:
            break
    return result
lst=[10,5,2,1]
amount=57
min_coins(lst,amount)        
        
```

```python
def activity_Selection(tasks):
    tasks.sort(key=lambda x:x[1])
    selected_tasks=[]
    selected_tasks.append(tasks[0])
    finish_time=tasks[0][1]
    for i in range(1,len(tasks)):
        if tasks[i][0]>=finish_time:
            selected_tasks.append(tasks[i])
            finish_time=tasks[i][0]

    return len(selected_tasks)
lst=[(1,2),(3,4),(0,6),(5,7),(8,9),(5,9)]
activity_Selection(lst)           
            
        
        
    
    
    
```
