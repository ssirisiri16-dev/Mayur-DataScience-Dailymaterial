```python
def covert_single(x):
    lst=x.split("-")
    if len(lst)==2:
        return (float(lst[0])+float(lst[1]))/2
    try:
        return float(x)
    except:
        return np.nan
    
```

```python
df.drop(columns=["sq_feet"],inplace=True)
```

```python
df
```

```python
df["new_sqft"]=df["total_sqft"].apply(covert_single)
```
