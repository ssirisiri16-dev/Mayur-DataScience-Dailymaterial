```python
import numpy as np
```

```python
def stimulate_bootstrapping(data,sample_size,no_samples,CI):
    sample_medians=[]
    for i in range(no_samples):
        sample=np.random.choice(data,size=sample_size)
        sample_medians.append(np.median(sample))
    sample_medians=np.array(sample_medians)
    sorted_medians=np.sort(sample_medians)
    per=np.percentile(sorted_medians,CI)
    return (sorted_medians[0], per)       
```

```python
data=np.random.randn(100)
sample_size=30
no_samples=1000
CI=95
stimulate_bootstrapping(data,sample_size,no_samples,CI)
```
