```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=np.array([1,2,3,4,5])
y=np.array([1,4,9,16,25])
```

```python
#name given to x-axis
plt.xlabel("Number")
#name given to y axis
plt.ylabel("squares")
#title given to the Graph
plt.title("Graph between x and y")
#function to plot the graph
plt.plot(x,y)
#it is function to show the graph
plt.show()
```

```python
p=np.array([2,3,4,5,6,7,8])
q=np.sqrt(p)
```

```python
#name given to x-axis
plt.xlabel("Number")
#name given to y axis
plt.ylabel("squaroot")
#title given to the Graph
plt.title("Graph between x and squareroot(x)")
#function to plot the graph
plt.plot(p,q)
#it is function to show the graph
plt.show()
```

```python
a=np.array([2,3,4,5,6,7,8])
b=np.log(a)
```

```python
#name given to x-axis
plt.xlabel("Number")
#name given to y axis
plt.ylabel("log")
#title given to the Graph
plt.title("Graph between x and log(x)")
#function to plot the graph
plt.plot(a,b)
#it is function to show the graph
plt.show()
```
