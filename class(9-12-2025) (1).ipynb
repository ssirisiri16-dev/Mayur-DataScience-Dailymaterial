```python
#22.Add a column is_final which marks 1 if the match was a final (based on match_id of last match in each season).
df1.columns
```

```python
#24.Create a new column run_type that classifies each ball as “boundary”, “dot”, or “other”.
df1.columns
```

```python
df2.columns
```

```python
df2["total_runs"].unique()
```

```python
def categorical_column(x):
    if x in [4,6,5,7]:
        return "Boundary"
    elif x==0:
        return "Dot"
    else:
        return "other"
```

```python
df2["run_type"]=df2["total_runs"].apply(categorical_column)
```

```python
df2
```
