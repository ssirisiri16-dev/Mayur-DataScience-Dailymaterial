```python
def job_sequencing_deadlines(profit,deadline):
    lst=[]
    noj=0
    profit=0
    for i in range(0,len(deadline)):
        lst.append((profit[i],deadline[i]))
    lst=sorted(lst,reverse=True)
    next_deadline=max(deadline)
    for i in range(0,len(lst)):
        if next_deadline-lst[i][1]>0:
            profit+=lst[i][0]
            noj+=1
    return noj,profit
    
```

```python
profit= [20, 10, 40, 30]
deadline=[4, 1, 1, 1]
job_sequencing_deadlines(profit,deadline)
```

```python
def jobSequencing(deadline, profit):
    n = len(profit)
    jobs = []
    for i in range(n):
        jobs.append((profit[i], deadline[i]))
    jobs.sort(reverse=True)  
    max_deadline = max(deadline)
    slots = [False] * max_deadline
    job_count = 0
    total_profit = 0
    for p, d in jobs:
        for j in range(d - 1, -1, -1):
            if not slots[j]:
                slots[j] = True
                job_count += 1
                total_profit += p
                break
    return [job_count, total_profit]
profit= [20, 10, 40, 30]
deadline=[4, 1, 1, 1]
jobSequencing(deadline, profit)
```

```python
def minimum_no_platform(arrival,departure):
    lst=[]
    count=1
    for i in range(0,len(departure)):
        lst.append((arrival[i],departure[i]))
    lst.sort(key=lambda x:x[0],reverse=True)
    for j in range(0,len(lst)-1):
        if lst[j+1][0]<=lst[j][1]:
            count+=1
    return count
arrival=[1000, 935, 1100]
departure = [1200, 1240, 1130]
minimum_no_platform(arrival,departure)   
```
