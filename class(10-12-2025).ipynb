```python
#30.Compute total runs scored by each team in each season.
df1.columns
```

```python
df2.columns
```

```python
data1=df1[["id","season","team1","team2"]]
```

```python
data2=df2[["match_id","total_runs"]]
```

```python
result=pd.merge(data1,data2,left_on="id",right_on="match_id",how="inner")
```

```python
result
```

```python
result.groupby(["season","team1","team2"])["total_runs"].sum()
```

```python
data3=df1[["id","season"]]
data4=df2[["match_id","batting_team","total_runs"]]
```

```python
res=pd.merge(data3,data4,left_on="id",right_on="match_id",how="inner")
```

```python
res.groupby(["season","batting_team"])["total_runs"].sum()
```

```python
#31.Calculate total wickets taken by each bowler across all seasons.
data6=df1[["id","season"]]
data7=df2[["match_id","bowler","is_wicket"]]
```

```python
res1=pd.merge(data6,data7,left_on="id",right_on="match_id",how="inner")
```

```python
res1.groupby(["season","bowler"])["is_wicket"].sum()
```

```python
#33.Get the average runs per over for each team in the 2017 season.
data8=df1[["id","season"]]
data9=df2[["match_id","over","batting_team","total_runs"]]
```

```python
res2=pd.merge(data8,data9,left_on="id",right_on="match_id",how="inner")
```

```python
res=res2[res2["season"]=="2017"]
```

```python
res
```

```python
res.groupby(["over","batting_team"])["total_runs"].mean()
```
